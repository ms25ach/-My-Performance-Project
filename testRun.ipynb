```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python

df = pd.read_csv("Wimbledon-men-2013.csv")
data = df[["FSW.1", "SSW.1"]].dropna()


x = data["FSW.1"]
y = data["SSW.1"]
```

```python
df.head()   
```

```python
#test case 1
plt.figure(figsize=(8,6))
plt.scatter(x, y)
slope, intercept, r, p, std_err = stats.linregress(x, y)
plt.plot(x, slope*x + intercept)
plt.title("Pearson Test: FSW.1 vs SSW.1")
plt.xlabel("First Serve Won (FSW.1)")
plt.ylabel("Second Serve Won (SSW.1)")
plt.grid(True)
plt.show()
```

```python
# 2. Spearman Rank Scatter Plot
x_rank = x.rank()
y_rank = y.rank()

plt.figure(figsize=(8,6))
plt.scatter(x_rank, y_rank)
plt.title("Spearman Rank Plot: FSW.1 vs SSW.1")
plt.xlabel("FSW.1 (Rank)")
plt.ylabel("SSW.1 (Rank)")
plt.grid(True)
plt.show()
```
